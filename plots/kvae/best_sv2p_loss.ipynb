```python
loss_path = '../../logs/KVAE/finetuned1/KVAE_scale=0.3_100.log'
```

```python
losses = []
recon_losses = []
log_qa = []
elbo_kf = []
mse = []

with open(loss_path) as f: 
    lines = f.readlines()
    for line in lines: 
        if "Namespace" in line or "Starting" in line or "Train" in line or "Finished" in line or "Saved" in line: 
                pass
        else: 
            line = line.strip("\n")
            line = line.strip("INFO:root:")
            
            line = line.split(",")
            losses.append(float(line[0]))
            recon_losses.append(float(line[1]))
            log_qa.append(float(line[2]))
            elbo_kf.append(float(line[3]))
            mse.append(float(line[4]))
        
```

```python
plt.xlabel('Epochs')
plt.title('Total Losses')

plt.plot(losses)
```

```python
plt.plot(recon_losses)

plt.xlabel('Epochs')
plt.title('Reconstruction Losses (Bernoulli)')
```

```python
plt.plot(log_qa)

plt.xlabel('Epochs')
plt.title('log q(a)')
```

```python
plt.plot(elbo_kf)

plt.xlabel('ELBO Kalman Filter')
plt.title('log q(a)')
```

```python
plt.plot(mse)
plt.yscale("log")

plt.xlabel('Epochs')
plt.title('MSE')
```
